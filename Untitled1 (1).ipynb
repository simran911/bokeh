Q1. How can you create a Bokeh plot using Python code?

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure,show

x=[1,2,3,4,5]
y=[6,7,8,9,10]

p=figure(title='Scatter plot',x_axis_label='x axis',y_axis_label='y axis')
p.circle(x,y,fill_color='pink',size=20)
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
from bokeh.plotting import figure ,show

x=[1,2,3,4,5]
y=[7,8,9,10,11]
fig=figure(title="Graph",x_axis_label="x axis",y_axis_label="y axis")
fig.triangle(x,y,color='red',legend_label="red triangle",size=20)
show(fig)

```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?



A Bokeh server is a Python library that allows you to create interactive web-based visualizations and applications. It provides a server-side infrastructure for generating and updating plots, charts, and other visual elements in real time. With Bokeh server, you can build interactive plots that respond to user inputs, update data dynamically, and enable real-time streaming of data.

To use Bokeh server and create interactive plots that can be updated in real time, you would typically follow these steps:

Install Bokeh: Start by installing the Bokeh library using pip or conda.

Import the necessary modules: In your Python script or Jupyter Notebook, import the required Bokeh modules, including bokeh.plotting or bokeh.models, depending on your specific needs.

Create a Bokeh plot: Use the Bokeh API to create your initial plot, specifying the type of plot, data, and visual attributes such as colors, markers, and labels.

Define callbacks: Next, define callback functions that will be triggered when certain events occur, such as a button click or a slider value change. These callbacks can update the data or properties of the plot dynamically.

Create a Bokeh server application: To serve the plot and make it interactive, create a Bokeh server application. This involves using the bokeh.server module and defining a curdoc object that represents the current document.

Add plot and callbacks to the application: Add your plot and callbacks to the Bokeh server application by modifying the curdoc object. This associates the plot with the server and allows it to be updated in real time.

Run the Bokeh server: Finally, run the Bokeh server using the bokeh serve command in the terminal or by running the script directly. This starts the server and makes the plot accessible through a web browser.

Once the Bokeh server is running, you can interact with the plot through the web browser. Any changes triggered by user inputs or dynamic data updates will be automatically reflected in the plot in real time.

By using the Bokeh server, you can create powerful interactive visualizations and applications that enable real-time updates and responsiveness, making it an excellent choice for data exploration, monitoring systems, and building interactive dashboards

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

To embed a Bokeh plot into a web page or dashboard using Flask or Django, you need to follow these general steps:

Set up your web framework: Install Flask or Django and create a new project or application.

Create a Bokeh plot: Use the Bokeh library to create the plot you want to embed. This involves defining the plot type, adding data, and configuring visual attributes.

Generate the Bokeh plot HTML: Bokeh provides a function called bokeh.embed.components() that generates the necessary HTML and JavaScript code to embed the plot in a web page. This function returns a tuple containing the HTML <div> element and the JavaScript code.

Integrate the Bokeh plot into your web framework:

Flask: In Flask, you can render a template that includes the Bokeh plot's HTML and JavaScript code. Pass the HTML <div> element and JavaScript code as variables to the template and use them to generate the final HTML page.

Django: In Django, you can create a view that generates the Bokeh plot's HTML and JavaScript code. Then, pass the HTML and JavaScript code as context variables to the template and render it to generate the final HTML page.

Serve the web page: Run your Flask or Django application to serve the web page that includes the embedded Bokeh plot. Access the page in a web browser to view the embedded plot.

Here's an example code snippet for embedding a Bokeh plot in Flask:

python
Copy code
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # Create a Bokeh plot
    p = figure(plot_width=400, plot_height=400)
    p.circle([1, 2, 3], [4, 5, 6])

    # Generate Bokeh plot HTML and JavaScript components
    script, div = components(p)

    # Render the template with the embedded plot
    return render_template('index.html', plot_script=script, plot_div=div)

if __name__ == '__main__':
    app.run()
In the above example, the components() function is used to generate the HTML and JavaScript components of the Bokeh plot. Then, these components are passed as variables to the render_template() function, which renders an HTML template (index.html) with placeholders for the plot's script and div. The rendered template is returned as the response for the / route.

In the index.html template, you would include the placeholders for the plot's script and div using Jinja2 templating syntax. For example:

The process for embedding a Bokeh plot in Django is similar, but you would define a Django view that generates the plot's HTML and JavaScript code, and then pass it to a template for rendering.

With these steps, you can embed Bokeh plots into your Flask or Django web applications and create interactive dashboards or visualizations.
